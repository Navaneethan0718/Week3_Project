```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("weatherHistory.csv")

print(df.head())
print(df.isnull().sum())

df.fillna(method='ffill', inplace=True)

df['Formatted Date'] = pd.to_datetime(df['Formatted Date'])
df['Month'] = df['Formatted Date'].dt.month

plt.figure()
plt.plot(df['Formatted Date'], df['Temperature (C)'])
plt.title("Temperature Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.show()

precip_counts = df['Precip Type'].value_counts()

plt.figure()
precip_counts.plot(kind='bar')
plt.title("Precipitation Type Comparison")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()

plt.figure()
plt.hist(df['Temperature (C)'], bins=30)
plt.title("Temperature Distribution")
plt.xlabel("Temperature (C)")
plt.ylabel("Frequency")
plt.show()

plt.figure()
plt.scatter(df['Temperature (C)'], df['Humidity'])
plt.title("Temperature vs Humidity")
plt.xlabel("Temperature (C)")
plt.ylabel("Humidity")
plt.show()

summary_counts = df['Summary'].value_counts().head(5)

plt.figure()
plt.pie(summary_counts, labels=summary_counts.index, autopct='%1.1f%%')
plt.title("Weather Summary Distribution")
plt.show()
```
